## 模範解答

- Li-Fraumeni症候群はなぜ常染色体顕性遺伝なのか。

  Li-Fraumeni症候群は、p53遺伝子のミスセンス変異によって悪性腫瘍に罹患しやすくなる疾患である。遺伝形式が常染色体顕性遺伝形式をとる理由は、p53がホモ4量体に変異型p53が1つでも入り込むと部分的、または全体的にがん抑制遺伝子としての機能を失うからである。

- スプライシングの分子生物学的機序を説明してください。

  スプライシングはスプライス部位の配列とsnRNPの相互作用によって進行する。スプライシングの過程は5’スプライス部位配列にU1-snRNP、ブランチ部位配列に分岐点結合タンパク(BBP)とU2補助因子(U2AF)が結合する。次の過程でBBPとU2AFがU2-snRNPと交代する。その後、U4-snRNPとU6-snRNPの複合体やU5-snRNPが結合した後で5’スプライス部位のC’5-OHがブランチ部位のアデニンのC’2に結合することによってラリアット構造を形成する。U1-snRNPやU4-snRNPが解離した後に、今度は5’スプライス部位のC’3-OHが3’スプライス部位のC’5を攻撃することによってイントロンとエキソンが完全に分離する。

- DNA複製時になぜトポイソメラーゼが必要なのか。

  DNA二重らせんは巻き不足であり、その代償として負のスーパーコイルを巻いている。複製を進めていくにつれて、DNAの巻きが緩められた部分ができる分だけ巻きがきつい部分ができてしまう。実際にはこの巻きがきつい部分は普通、正のスーパーコイルとして現れる。つまり、複製を完全に進めるためには正のスーパーコイルを弛緩するような酵素が必要である。その酵素がトポイソメラーゼである。

- エピジェネティクスの一例としてヒストン修飾を説明せよ。

  ヒストンは普段はDNAと静電気的に固く結びついているために、ヒストンとDNAとの相互作用の強さは転写の頻度に大きな影響を与える。ヒストンの修飾はそれらの相互作用を変化させる働きを持っている。一例として、リジン残基のアセチル化は、転写頻度を上げる役割がある。DNAとの静電気的相互作用を減弱させることで基本転写因子がアクセスしやすくしたり、ブロモドメインを持つ基本転写因子やコアクチベーターなどがアセチル化リジンを持つヒストンに結合したりするからである。

- ビスホスホグリセリン酸(BPG)の低酸素適応への役割を「アロステリック効果」と「酸素解離曲線」という言葉を用いて説明せよ。

  2,3-ビスホスホグリセリン酸は、低酸素状態下において解凍系から産生される。2,3-ビスホスホグリセリン酸はヘモグロビンと結合し、アロステリック効果により酸素親和性を低下させる。酸素解離曲線は右方移動し、低酸素化下において結合酸素が減少させることで、末梢における酸素供給を高める。

# 1. ベースラインモデル評価
**まずはベースモデルがどの程度知識を持っているか確かめる**

```python
def generate_output(query, system_prompt=None):
  if system_prompt is None:
    messages = [
        {"role": "user", "content": query},
    ]
  else:
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": query},
    ]
  input_ids = tokenizer.apply_chat_template(
      messages,
      add_generation_prompt=True,
      return_tensors="pt"
  ).to(model.device)

  terminators = [
      tokenizer.eos_token_id,
      tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = model.generate(
      input_ids,
      max_new_tokens=512,
      eos_token_id=terminators,
      do_sample=False,
      # temperature=0.6, # If do_sample=True
      # top_p=0.9,  # If do_sample=True
  )

  response = outputs[0][input_ids.shape[-1]:]
  return tokenizer.decode(response, skip_special_tokens=True)
```

```python
system_prompt = "質問に回答してください。必ず「日本語で回答」すること。"
question =  "Li-Fraumeni症候群はなぜ常染色体顕性遺伝なのか。"
response = generate_output(question, system_prompt)
```

# 2. RAGの導入


* **分割方法**: 「。」（句点）で区切られた文単位でテキストを分割
* **検索手法**: シンプルな類似度ベースの検索でクエリに関連する文を抽出
* **制約条件**: モデルの入力トークン制限に収まるよう関連文のみを選択

```python
from sentence_transformers import SentenceTransformer

emb_model = SentenceTransformer("infly/inf-retriever-v1-1.5b", trust_remote_code=True)
# In case you want to reduce the maximum length:
emb_model.max_seq_length = 8192
```

```python
with open("/content/lecture-ai-engineering/day3/data/Biochemo_raw.txt", "r") as f:
  raw_writedown = f.read()
```

```python
# ドキュメントを用意する。
documents = [text.strip() for text in raw_writedown.split("。")]
print("ドキュメントサイズ: ", len(documents))
print("ドキュメントの例: \n", documents[250])
```

```python
# Retrievalの実行
question = "Li-Fraumeni症候群はなぜ常染色体顕性遺伝なのか。"

query_embeddings = emb_model.encode([question], prompt_name="query")
document_embeddings = emb_model.encode(documents)

# 各ドキュメントの類似度スコア
scores = (query_embeddings @ document_embeddings.T) * 100
print(scores.tolist())
```

```python
topk = 5
for i, index in enumerate(scores.argsort()[0][::-1][:topk]):
  print(f"取得したドキュメント{i+1}: (Score: {scores[0][index]})")
  print(documents[index], "\n\n")
```

```python
references = "\n".join(["* " + documents[i] for i in scores.argsort()[0][::-1][:topk]])
system_prompt = "質問に回答してください。必ず「日本語で回答」すること。また、与えられる資料を参考にして回答すること。"
question =  f"[参考資料]\n{references}\n\n[質問] Li-Fraumeni症候群はなぜ常染色体顕性遺伝なのか。"
response = generate_output(question, system_prompt)
```

## 考察

- RAGの有効性

RAGを導入する前は部分的に誤り2件、説明不十分1件、誤答2件であったが、RAGを導入することで正答1件、部分的に誤り2件、誤答2件と改善した。また、部分的に誤りや誤答はより正答に近づいている。例えば、1問目ではLi-Fraumeni症候群がTP53遺伝子異常による疾患であることを理解した回答になったり、5問目では酸素解離曲線が右方移動することをRAG導入後は理解していた。

- RAGの限界

RAGを導入したことで、参照文章を参考にするあまりに、余計な部分までも書いてしまうことがあった。1問目の「1つの染色体上に異常があるだけでなく、もう1つの染色体上でも異常がある場合に、がんが生じるという考え方である」や5問目の「このアロステリック効果は、水素イオン濃度に依存してヘモグロビンの酸素親和性を変化させることを示しています。」の部分は回答に関係ない部分である。回答に関係ない部分をうまく認識する技術の必要性を理解した。

また、今回資料をあえて訂正しなかったのだが、本来、顕性遺伝と優性遺伝は同義であるが、参照文書では優性、質問文では顕性という言葉を用いたためモデルが混乱した。時間が進み、知識がアップデートされる中でRAGの参照文書もアップデートしていかなければならないと考えた。
